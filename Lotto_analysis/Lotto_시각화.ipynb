### 숫자별 hist

```python
fig, axes = plt.subplots(7,1, figsize=(16,42))
sns.histplot(data=lotto["a"], bins=45, ax=axes[0]).set_title("1번숫자")
sns.histplot(data=lotto["b"], bins=45, ax=axes[1]).set_title("2번숫자")
sns.histplot(data=lotto["c"], bins=45, ax=axes[2]).set_title("3번숫자")
sns.histplot(data=lotto["d"], bins=45, ax=axes[3]).set_title("4번숫자")
sns.histplot(data=lotto["e"], bins=45, ax=axes[4]).set_title("5번숫자")
sns.histplot(data=lotto["f"], bins=45, ax=axes[5]).set_title("6번숫자")
sns.histplot(data=lotto["bonus"], bins=45, ax=axes[6]).set_title("보너스숫자");
```

### 전체, 최근100, 최근 50의 번호 평균

```python
lotto_mean = lotto.describe().loc[["mean"]]
lotto_mean = pd.concat([lotto_mean, lotto_100, lotto_50])
lotto_mean = lotto_mean.reset_index()
lotto_mean = lotto_mean[["a", "b", "c", "d", "e", "f", "bonus"]]
lotto_mean
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=lotto_mean)
```

### 각 숫자별 나온 횟수

```python
count_num = pd.concat([lotto["a"], lotto["b"], lotto["c"], lotto["d"], lotto["e"], lotto["f"]]).value_counts().sort_index()
```

```python
count_num.mean()
```

```python
count_num
```

```python
count_num.describe()
```

```python
count_num.plot()
plt.plot((1,45), (141,141), "red")
plt.show()
```

```python
count_num.nlargest(6)
```

```python
count_num.nsmallest(6)
```
